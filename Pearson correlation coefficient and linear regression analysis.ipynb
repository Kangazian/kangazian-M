```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
```

```python
#Part one 1
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

Std_x = np.std(x, ddof=1) 
Std_y = np.std(y, ddof=1)
```

```python
cov_xy = np.cov(x, y, ddof=1)[0, 1]
r = cov_xy / (Std_x * Std_y)
b_1 = r * (Std_y / Std_x)
print('Pearson C',r)
print('Slope of Regression',b_1)
```

```python
#Part two 2
r,n=0.9,30
t_statistic=(r * np.sqrt(n - 2)) / np.sqrt(1 - r**2)
degrees_of_freedom = n - 2
print('t',t_statistic)
print('Degrees of freedom',degrees_of_freedom)
```

```python
#Part three 3
def critical_r(n, confidence_level):
    df = n - 2
    t_crit = t.ppf((1 + confidence_level) / 2, df)
    r_crit = t_crit / np.sqrt(t_crit**2 + df)
    return r_crit
```

```python
sample_sizes = np.arange(3, 251)
confidence_intervals = [0.99, 0.95, 0.90]
colors = ['r', 'g', 'b']
labels = ['99%', '95%', '90%']
```

```python
plt.figure(figsize=(10, 6))
for ci, color, label in zip(confidence_intervals, colors, labels):
    critical_values = [critical_r(n, ci) for n in sample_sizes]
    plt.plot(sample_sizes, critical_values, label=f'CI {label}', color=color)

plt.xlabel('Number of Sample')
plt.ylabel('Critical Value Of Pearson Correlation')
plt.title('Critical Values of pearson correlation vs sample size')
plt.legend()
plt.grid(True)
plt.show()
```
